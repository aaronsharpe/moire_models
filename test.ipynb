```python
from lattice_model import *
import matplotlib.pyplot as plt
```

```python
side_length = 0.5
scale = 1.8
lattice_vec_0 = scale*np.array([side_length*1.5, np.sqrt(3)*side_length/2])
lattice_vec_1 = scale*np.array([side_length*1.5,-np.sqrt(3)*side_length/2])
lattice_vecs = [lattice_vec_0, lattice_vec_1]

hex = gen_hexagon(side_length, 0, 0)

polys = gen_lattice_circ(hex, lattice_vecs)
```

```python
plt.figure()
for poly in polys:
    plt.plot(poly[:, 0], poly[:, 1])
```
